```python
import pandas as pd

df=pd.read_excel('Cadeiras gamers.xlsx')
df
```

```python
cores = ['branco','preto','vermelho','verde','dourado','cinza','roxo',
         'rosa','azul','laranja','creme',
         'marrom','preta','yellow','blue','gray',
         'purple','pink','off white','all black','orange','bronw','gold','red','green','black','white','branca']

def extrair_cores(texto):
    texto = texto.lower()
    for cor in cores:
        if cor in texto:
            return cor
    return 'Desconhecido'
 

df["cor_principal"] = df['marca'].apply(extrair_cores)

df[['marca', 'cor_principal']].head(10)





```

```python
cores_presentes = df['cor_principal'].unique()


cores_faltando = [cor for cor in cores if cor not in cores_presentes]

print("Cores faltando:", cores_faltando)
print("Quantidade de cores faltando:", len(cores_faltando))
```
