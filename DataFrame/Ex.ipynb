```python
import pandas as pd
import numpy as np
df= pd.read_excel('vendas.xlsx')
```

```python
#1.Padronizar a coluna categoria para minúsculas.

df['categoria'] = df['categoria'].str.lower()
df

```

```python
#2.Remover registros onde a quantidade está ausente
df.dropna(subset=['quantidade'])
df
```

```python
#3. Remover espaços extras e padronizar com primeira letra maiúscula (estilo título)

df['categoria']= df['categoria'].str.title()
df['loja']= df['loja'].str.title()

df
```

```python
#4. Garantir que a coluna data seja datetime

df['data']=pd.to_datetime(df['data'])
df
```

```python
#5. Calcular valor unitário real das vendas válidas

#df['valor_unitario'] = df['valor_venda'] / df['quantidade']
df
```

```python
#6. Calcular valor médio unitário por produto (sem considerar nulos)

media_uni =df.groupby('produto')['valor_unitario'].mean()
media_uni
```

```python
#7. Agrupar e calcular média
df_agrupado= df.groupby('valor_venda')['valor_unitario'].mean()
df
```

```python
#8. Preencher valor_unitario ausente

df['valor_unitario'].fillna(df['valor_unitario'].mean(),inplace=True)
df
```

```python
#9. Recalcular valor_venda
df["valor_recalque"] = df["valor_unitario"].groupby(df["produto"]).transform("mean")
df["valor_venda"] = df["quantidade"] * df["valor_recalque"]
df["valor_venda"]
```

```python
#10. Salvar o novo DataFrame limpo em um novo arquivo Excel

df.to_excel('vendanew.xlsx',index=False)
df
```

```python
#1. Total de vendas por loja - Agrupar por loja e somar o valor total vendido

df.groupby('loja')['venda_recalque'].sum()
df


```

```python
#2. Evolução mensal de vendas -  Identificar sazonalidades e picos mensais.
import matplotlib.pyplot as plt

df['mes_ano']=df['data'].dt.to_period('M')
vendas_mensais = df.groupby('mes_ano')['valor_venda'].sum()
vendas_mensais

plt.figure(figsize=(10, 6))
vendas_mensais.plot(kind='line', marker='o')
plt.title('Evolução Mensal de Vendas')
plt.xlabel('Mês/Ano')
plt.ylabel('Total de Vendas')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()


media_anual = vendas_mensais.mean()


meses_acima_media = vendas_mensais[vendas_mensais > media_anual]
meses_abaixo_media = vendas_mensais[vendas_mensais < media_anual]

print(f'Meses com vendas acima da média:\n{meses_acima_media}')
print(f'Meses com vendas abaixo da média:\n{meses_abaixo_media}')

pico_maximo = vendas_mensais.idxmax()
pico_minimo = vendas_mensais.idxmin()

print(f'Mês com maior pico de vendas: {pico_maximo} com {vendas_mensais[pico_maximo]} vendas')
print(f'Mês com menor pico de vendas: {pico_minimo} com {vendas_mensais[pico_minimo]} vendas')

```

```python
#3. Agrupar por mês e somar vendas

vendas_mensais = df.groupby('mes_ano')['valor_venda'].sum()
vendas_mensais



```

```python
#4. Produtos mais vendidos (por volume) Objetivo: saber o que mais sai no caixa.
df['quantidade_vendida'] = df['valor_venda'] / df['valor_unitario']
produtos_mais_vendidos = df.groupby('produto')['quantidade'].sum().reset_index()
produtos_mais_vendidos = produtos_mais_vendidos.sort_values(by='quantidade', ascending=False)
produtos_mais_vendidos
```

```python
#5. Comparação entre categorias. Objetivo: analisar quais categorias são mais lucrativas.
lucro_por_categoria = df.groupby('categoria')['valor_venda'].sum().reset_index()
lucro_por_categoria = lucro_por_categoria.sort_values(by='valor_venda', ascending=False)
lucro_por_categoria
```

```python
#6. Comparação entre lojas por categoria. Objetivo: avaliar se há especialização ou foco diferente entre as lojas, com base nas categorias que mais vendem. Agrupar por loja e categoria e calcular o lucro total
resultado = df.groupby(['loja', 'categoria'])['valor_venda'].sum().reset_index()

resultado_pivot = resultado.pivot(index='loja', columns='categoria', values='valor_venda').fillna(0)


print("Lucro por loja e categoria:")
resultado
```

```python
#7. Plotar como gráfico de barras empilhadas
import pandas as pd
import matplotlib.pyplot as plt



resultado_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))


plt.title('Vendas por Loja e Categoria')
plt.xlabel('Lojas')
plt.ylabel('Valor de Vendas')
plt.legend(title='Categoria')
plt.xticks(rotation=45)
plt.tight_layout()  


plt.show()
```
